```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict, Any
from dotenv import load_dotenv
load_dotenv()
from langchain_groq import ChatGroq
```

```python
llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.0
)
```

```python
class State(TypedDict):
    topic : str
    outline : str
    blog : str
```

```python
# define graph
graph = StateGraph(State)
```

```python
#define node functions
def outline_node(state: State) -> State:

    # fetch topic from state and create prompt
    topic=state['topic']

    # create prompt and invoke llm to get outline
    prompt = f"Write a detailed outline for a blog post about - {topic}."

    # invoke llm and save outline to state
    outline = llm.invoke(prompt).content
    state['outline'] = outline

    #return state
    return state
```

```python
def blog_node(state: State) -> State:

    # fetch outline from state and create prompt
    outline = state['outline']

    # create prompt and invoke llm to get blog post
    prompt = f"Write a detailed blog post based on the following outline - {outline}."

    # invoke llm and save blog post to state
    blog = llm.invoke(prompt).content
    state['blog'] = blog
    
    return state
```

```python
# nodes
graph.add_node("outline_node", outline_node)
graph.add_node("blog_node", blog_node)
```

```python
# edges
graph.add_edge(START, "outline_node")
graph.add_edge("outline_node", "blog_node")
graph.add_edge("blog_node", END)
```

```python
# compile graph into a workflow
workflow = graph.compile()
```

```python
initial_state = {'topic':'The future of AI Agents'}
final_state = workflow.invoke(initial_state)
print(final_state['blog'])
```
