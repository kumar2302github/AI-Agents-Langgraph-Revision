```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
class State(TypedDict):
    question : str
    answer : str
```

```python
# define graph
qa_graph = StateGraph(State)
```

```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(
    model="llama-3.1-8b-instant",
    temperature=0.0,
    max_retries=2,
    # other params...
)
```

```python
#define nodes 
def llm_qa(state: State) -> State:

    # extract the question from the state
    question =State["question"]

    # form a prompt 
    prompt = f"Answer the question: {question}"

    # invoke the LLM
    answer = llm.invoke(prompt).content 

    # update the state with the answer

    State['answer'] = answer

    return State
```

```python
# nodes and edges
qa_graph.add_node("llm_qa", llm_qa)

qa_graph.add_edge(START, "llm_qa")
qa_graph.add_edge("llm_qa", END)
```

```python
workflow = qa_graph.compile()
```

```python
initial_state = {"question":"What is the capital of France?"}
final_state = workflow.invoke(initial_state)
print(final_state) #final_state
```
